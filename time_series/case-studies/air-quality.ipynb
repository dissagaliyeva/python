```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

sns.set()
matplotlib.rcParams['figure.figsize'] = (17, 5)

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data/AirQualityUCI.csv', sep=';', decimal=',')
data.info()
```

```python
data.isna().sum()
```

```python
data.drop(['Unnamed: 15', 'Unnamed: 16'], axis='columns', inplace=True)
```

```python
# drop nan values
data = data[data['Date'].notnull()]
data.isna().sum()
```

```python
data['Datetime'] = data['Date'] + ' ' + data.Time
data.drop(['Date', 'Time'], axis='columns', inplace=True)
data['Datetime'] = pd.to_datetime(data['Datetime'], format='%d/%m/%Y %H.%M.%S')
data.info()
```

```python
data.index = data.Datetime

data[['T']].plot()
```
