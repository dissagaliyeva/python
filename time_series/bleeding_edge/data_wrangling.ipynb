## Resampling

Resampling is a method to convert the time-based observed data into a different interval. In other words, it is used to change the time frequency into another time-frequency format. For instance, say we want to change monthly data into a year-wise format or upsample week data into hours. We would perform either upsample or downsample operations.

```python
birth_df = pd.read_csv('https://raw.githubusercontent.com/Apress/hands-on-time-series-analylsis-python/master/Data/daily-total-female-births-CA.csv', parse_dates=True, index_col=0)
birth_df.head(3)
```

```python
# resample by month
birth_df.births.resample('M').mean().plot()
```

```python
# resample by quarter
birth_df.births.resample('Q').mean().plot()
```

```python
# resample by weekly
birth_df.births.resample('W').mean().plot()
```

```python
# resample on a semi-monthly basis (twice/month)
birth_df.births.resample('SM').mean().plot()
```

## Windowing functions

### Rolling window

Rolling window is useful when we want to calculate a small range of values and the results (just like a moving range slider). For example, if we have 20 days temperature, and we want to calculate the differences within 3 days and keep track of them - that's the rolling window.

The rolling window feature supports the following methods: count, sum, mean, median, var, std, min, max, corr, cov, skew, kurt, quantile, sum, and aggregate.

```python
istanbul_stocks = pd.read_excel('data/edge_techniques/istambul_stock_exchange.xlsx', parse_dates=['date'], index_col=0)
istanbul_stocks
```

```python
# rolling
rolling = istanbul_stocks.rolling(window=4).mean()
rolling
```

```python
rolling.plot()
```

### expanding

```python
expanding = istanbul_stocks.expanding(min_periods=4).mean()
expanding
```

```python
expanding.plot()
```

### exponentially weighted moving window

```python
ewm = istanbul_stocks.ewm(com=0.5).mean()
ewm
```

```python
ewm.plot()
```

### shifting

```python
shifted = istanbul_stocks.shift(periods=3)
shifted
```
